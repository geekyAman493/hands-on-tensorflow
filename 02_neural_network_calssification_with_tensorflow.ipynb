# to visualize model's predictions let's create a function
* take trained x and y
* create a numpy mesh grid for diff x values
* make predictions across the meshgrid
* plot the predictions as well as line between zone (where each unique class falls)

```python
import numpy as np
```

```python
def plot_decision_bndry(model,x,y):


  #define the axis boundaries of the plot and create a mesh grid
  x_min,x_max=x[:,0].min()-0.1,x[:,1].max()+0.1
  y_min,y_max=x[:,0].min()-0.1,x[:,1].max()+0.1

  xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),
                    np.linspace(y_min,y_max,100))

  #create x values
  x_in =np.c_[xx.ravel(), yy.ravel()] #stack 2d arrays together

  #make predictions
  y_pred= model.predict(x_in)

  #check for multiclass

  if len(y_pred[0])>1:
    print('doing multiclass classification')
    y_pred =np.argmax(y_pred,axis=1).reshape(xx.shape)
  else:
    print('doing binary')
    y_pred=np.round(y_pred).reshape(xx.shape)


    #plot decision boundary
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(x[:,0],x[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
#check our decision bndry
plot_decision_bndry(model=model_3,x=x,y=y)
```

The decision boundary is highly linear i.e a straight line where as our data is non linear circular

# Sigmoid Activation function `f(x)=1/(1+exp(-x))`

```python
#lets start by replicating sigmoid - sigmoid(x)=1/(1+exp (-x))
def sigmoid(x):
  return 1/(1+tf.exp(-x))


#use it on our toy tensor
sigmoid(a)
```

```python
#plot sigmoid(a)

plt.plot(sigmoid(a))
plt.title('sigmoid')
```

# Relu Activation function - `f(x)=max(x,0)`

```python
#lets recreate relu function

def relu(x):
  return tf.maximum(x,0)

relu(a)
```

```python
plt.plot(relu(a))
plt.title('relu')
```

# TanH Activation function - `tanh(x)=sinh(x)/cosh(x)` i.e
#  `(exp(x)-exp(-x))/(exp(x)+exp(-x))`



```python
def tanh(x):
  return (tf.exp(x)-tf.exp(-x))/(tf.exp(x)+tf.exp(-x))
```

```python
tanh(a)
```

```python
plt.plot(tanh(a))
plt.title('tanh')
```

# Finding the best learning rate
To find the ideal learning rate (the learning rate where the loss decreases the most during training) we're going to use the following steps :

* A `learning rate callback` - its a extra piece of function which can be added to model when training

* Another model

* A modified loss curves plot

```python
tf.random.set_seed(42)

#let's create a model same as model_7

model_8=tf.keras.Sequential([
    tf.keras.layers.Dense(5,activation='relu'),
    tf.keras.layers.Dense(5,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

#compile

model_8.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

#create a learning rate callback

lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

#fit
history_8=model_8.fit(x_train,
                      y_train,
                      epochs=100,
                      callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history_8.history).plot()
```

```python
lrs=1e-4*(10**(tf.range(100)/20))
plt.semilogx(lrs,history_8.history['loss'])
plt.xlabel('Learning rate')
plt.ylabel('Loss')
plt.title('learning rate vs loss')
```
